```python
from transformers import MarianMTModel, MarianTokenizer
```

```python
def eng_hindi_translate(text, target_language):
    # Define the model name
    model_name = f'Helsinki-NLP/opus-mt-en-{target_language}'
    
    # Instantiate the tokenizer
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    
    # Instantiate the model
    model = MarianMTModel.from_pretrained(model_name)
    
    # Tokenize input text
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    
    # Translate text
    outputs = model.generate(**inputs, num_beams=4, max_length=1000, early_stopping=True)
    
    # Decode translated text
    translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    
    return translated_text
```

```python
english_text = "Ram Navami is a Hindu festival celebrating the birth of Lord Rama, the seventh avatar of Vishnu. It falls on the ninth day of Chaitra month in the Hindu calendar. Devotees observe fasts, visit temples, and recite Ramayana scriptures to honor the ideals of righteousness and dharma associated with Lord Rama."
hindi_translation = eng_hindi_translate(english_text, "hi")
print(hindi_translation)
```
